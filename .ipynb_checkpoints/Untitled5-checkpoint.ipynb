```python
import nltk
import os
import re
import random
```

```python
from nltk.corpus.reader import CategorizedPlaintextCorpusReader
reader = CategorizedPlaintextCorpusReader('.\\amazon', r'a.*.txt', cat_pattern = r'a([a-z][a-z][a-z])\d+')
```

```python
reader[0]
```

```python
reader.categories('aneg0.txt')
```

```python
reader.fileids(0)
```

```python
documents = [(list(reader.words(fileid)), category)
             for category in reader.categories()
            for fileid in reader.fileids(category)]
```

```python
random.shuffle(documents)
```

```python
all_words = nltk.FreqDist(w.lower() for w in reader.words())
word_features = all_words.keys()
```

```python
def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains(%s)' % word] = (word in document_words)
    return features
```

```python
print document_features(reader.words('apos0.txt'))
```

```python
featuresets = [(document_features(d), c) for (d, c) in documents]
len(featuresets)
```

```python
train_set, test_set = featuresets[300:], featuresets[:300]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print nltk.classify.accuracy(classifier, test_set)
```

```python
classifier.show_most_informative_features(10)
```
